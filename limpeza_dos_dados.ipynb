```python
#importação das bibliotecas
import pandas as pd
import os
```

```python
#função para limpar os dados de cada arquivo csv anual e salvar como novo arquivo

def limpar_dados_incendio(nome_arquivo, novo_arquivo):
    df = pd.read_csv(nome_arquivo)
    df = df.drop(columns=["id_bdq", "foco_id"])
    df[["data", "hora"]] = df["data_pas"].str.split(" ", expand=True)
    df = df.drop(columns=["data_pas", "hora"])
    df[["ano", "mês", "dia"]] = df["data"].str.split("-", expand=True)
    df = df.drop(columns=["data", "pais"])
    df = df.drop_duplicates()

    df.to_csv(novo_arquivo, index=False)

    return df

df = limpar_dados_incendio("focos_br_ref_2024.csv", "2024_limpo.csv")
display(df)
```

Concatenção dos arquivos:

```python
#pasta onde os arquivos estão localizados
pasta = "."

#nome do arquivo csv final com os dados processados
arquivo_out = "casos_incendio.csv"

dataframes = []

#os.listdir para listar arquivos dentro da pasta
for arquivo in os.listdir(pasta):
    if arquivo.endswith("_limpo.csv"):
        caminho_arquivo = os.path.join(pasta,arquivo)
        df = pd.read_csv(caminho_arquivo)
        dataframes.append(df)
        
#para combinar os dataframes em um só
dados = pd.concat(dataframes, ignore_index=True)
dados = dados.sort_values(by=["ano","mês"], ascending=[False, True])
dados = dados.drop_duplicates()
dados.to_csv(arquivo_out, index=False)
dados
```
